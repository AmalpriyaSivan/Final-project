```python
import numpy as np
from sklearn.utils import resample
import matplotlib.pyplot as plt

# Assuming you have already defined the 'labels' array
classes, counts = np.unique(labels, return_counts=True)

# Plot the class distribution before downsampling
plt.bar(classes, counts, tick_label=['meningioma (0)', 'glioma (1)', 'pituitary tumor (2)', 'no_tumor (3)'])
plt.xlabel('Classes')
plt.ylabel('Number of Samples')
plt.title('Class Distribution Before Downsampling')
plt.show()

# Determine the size of the smallest class
minority_class_size = min(counts)

# Downsampling each class to match the size of the smallest class
downsampled_indices = []
for idx in classes:
    class_indices = np.where(labels == idx)[0]
    if idx == 3:  # Handling the 'no_tumor' class separately
        # For the 'no_tumor' class, we will resample with replacement to match the minority class size
        downsampled_indices.extend(resample(class_indices, replace=True, n_samples=minority_class_size, random_state=42))
    else:
        downsampled_indices.extend(resample(class_indices, replace=False, n_samples=minority_class_size, random_state=42))

# Collect the downsampled data based on the downsampled indices
features_downsampled = features[downsampled_indices]
labels_downsampled = labels[downsampled_indices]

# Now, the 'features_downsampled' and 'labels_downsampled' arrays contain the downsampled data.

# Plot the class distribution after downsampling
classes_downsampled, counts_downsampled = np.unique(labels_downsampled, return_counts=True)
plt.bar(classes_downsampled, counts_downsampled, tick_label=['meningioma (0)', 'glioma (1)', 'pituitary tumor (2)', 'no_tumor (3)'])
plt.xlabel('Classes')
plt.ylabel('Number of Samples')
plt.title('Class Distribution After Downsampling')
plt.show()

# Define the mapping from class integers to class names
integer_to_class = {
    '0': 'meningioma',
    '1': 'glioma',
    '2': 'pituitary tumor',
    '3': 'no_tumor'
}

# Print the number of samples per class after downsampling
for i, idx in enumerate(classes_downsampled):
    print('Number of {}: {}'.format(integer_to_class[str(idx)], counts_downsampled[i]))
```

```python
from skimage.transform import resize

# Create an empty array to store the resized images
resized_features = np.zeros((features_downsampled.shape[0], 512, 512))

# Iterate through each image and resize it to 512x512
for i in range(features_downsampled.shape[0]):
    resized_features[i] = resize(features_downsampled[i, 0], (512, 512))

# Now, the 'resized_features' array contains all the images resized to 512x512
```

```python
from skimage.filters import threshold_otsu

# Assuming you have 'resized_features' after downsampling

# Create an empty array to store the thresholded masks
tumor_masks = np.zeros_like(resized_features)

# Iterate through each image and apply Otsu's thresholding
for i in range(resized_features.shape[0]):
    # Compute the threshold value using Otsu's method
    threshold_value = threshold_otsu(resized_features[i])

    # Threshold the image to create the mask
    tumor_masks[i] = resized_features[i] > threshold_value
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming you have 'resized_features' after downsampling and 'tumor_masks' as the segmented tumor masks

# Create an empty array to store the overlayed images with RGB channels
overlayed_images = np.zeros((resized_features.shape[0], 512, 512, 3))

# Iterate through each image and overlay the tumor region on the original image
for i in range(resized_features.shape[0]):
    # Use the mask as a binary mask to extract the tumor region from the original image
    tumor_region = resized_features[i] * tumor_masks[i]

    # Create an RGB image with the same shape as the original image
    overlayed_image = np.stack([resized_features[i]] * 3, axis=-1)

    # Set the tumor region to red (255) in the overlayed image
    overlayed_image[tumor_region != 0] = [255, 0, 0]

    # Save the overlayed image
    overlayed_images[i] = overlayed_image

# Now, 'overlayed_images' contains the original images with the segmented tumor regions shown in red.

# Visualize the overlayed images along with original images and tumor masks
plt.figure(figsize=(16, 8))
for i, idx in enumerate(np.random.randint(resized_features.shape[0], size=18), start=1):
    plt.subplot(3, 6, i)

    # Extract the 2D grayscale image slice
    image_slice = resized_features[idx]
    plt.imshow(image_slice, cmap='gray')

    # set black pixel as transparent for the mask
    mask = np.ma.masked_where(tumor_masks[idx] == False, tumor_masks[idx])
    plt.imshow(mask, alpha=0.3, cmap='Set1')

    # Show the overlayed image
    plt.imshow(overlayed_images[idx], alpha=0.5)

    plt.title(integer_to_class[str(labels_downsampled[idx])])
    plt.axis('off')

plt.show()
```

```python
# Calculate the area of the tumor region in each overlayed image
tumor_areas = []

for i in range(resized_features.shape[0]):
    # Use the mask as a binary mask to extract the tumor region from the overlayed image
    tumor_region = overlayed_images[i][:, :, 0] == 255  # Check the red channel for the tumor region

    # Calculate the area of the tumor region by counting the non-zero pixels in the mask
    tumor_area = np.sum(tumor_region)

    tumor_areas.append(tumor_area)

# Print the tumor areas
print("Tumor Areas:")
for i, area in enumerate(tumor_areas, start=1):
    print(f"Image {i}: {area} pixels")
```
